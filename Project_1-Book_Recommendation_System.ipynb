```python
x=merged_data.groupby('User-ID').count()['Book-Rating']>200
Educated_Users=x[x]#Boolean Indexing
Educated_Users
```

```python
x=merged_data.groupby('User-ID').count()['Book-Rating']>200
Educated_Users=x[x].index
Educated_Users
```

```python
filtered_ratings=merged_data[merged_data['User-ID'].isin(Educated_Users)]
filtered_ratings
```

```python
num_ratings_df3=filtered_ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings_df3.rename(columns={'Book-Rating':'BookRatings-Count'},inplace=True)
num_ratings_df3
```

```python
num_ratings_df3[num_ratings_df3[['BookRatings-Count']]>50].boxplot()
```

```python
y=filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
famous_books
```

```python
final_ratings=filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def reccomend(book_name):
    #index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    data=[]
    for i in similar_items:
        item=[]
        temp_df=Books[Books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```

```python
reccomend('1984')
```
